```python
import pandas as pd

# Load the data and parse the 'date' column as datetime
trans = pd.read_csv('/content/transactions.csv',
                    parse_dates=True)

# Filter the data for store number 47
trans_47 = trans[trans['store_nbr'] == 47]

# Extract the 'year' and 'month' from the 'date' column
trans_47['date'] = pd.to_datetime(trans['date'])
trans_47['year'] = trans_47['date'].dt.year
trans_47['month'] = trans_47['date'].dt.month

# Aggregate by 'year' and 'month', summing the 'transactions'
trans_agg = trans_47.groupby(['year', 'month'])['transactions'].sum().reset_index()

# Add 'previous_year' column by shifting the 'transactions' column by 12 months
trans_agg['previous_year'] = trans_agg['transactions'].shift(12)

# Filter for year 2015 and drop the unnecessary columns, then set 'month' as the index
trans_agg_filtered = trans_agg[(trans_agg['year'] == 2015)]
trans_agg_filtered = trans_agg_filtered.drop(['year'], axis=1).set_index('month')

# Optionally, you can plot the result
trans_agg_filtered.plot()

# Print the resulting DataFrame
print(trans_agg_filtered)
```

```python
total['date'] = pd.to_datetime(total['date'])
grp = total.groupby(['date','type'])['sales'].sum().reset_index()
grp.pivot_table(index='date', columns='type', values='sales', aggfunc='sum').plot(cmap='Set2')
```

```python
total['month'] = total['date'].dt.month
total.pivot_table(
    index = 'month',
    columns = 'type',
    values = 'sales',
    aggfunc = 'sum'
).plot.bar(stacked=True)
```
