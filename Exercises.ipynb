```python
retail.query("family in ['AUTOMOTIVE', 'BEVERAGES', 'BEAUTY'] and date in ['2016-01-01', '2016-01-02','2016-01-03']").pivot_table(
    index = 'family',
    columns = 'date',
    values = 'sales',
    aggfunc = 'sum',
    margins = True
)
```

```python
retail.query("family in ['AUTOMOTIVE', 'BEVERAGES', 'BEAUTY'] and date in ['2016-01-01', '2016-01-02','2016-01-03']").pivot_table(
    index = 'family',
    columns = 'date',
    values = 'sales',
    aggfunc = ('min', 'max')
)
```

```python
retail.query("family in ['AUTOMOTIVE', 'BEVERAGES', 'BEAUTY'] and date in ['2016-01-01', '2016-01-02','2016-01-03']").pivot_table(
    index = 'family',
    columns = 'date',
    aggfunc = ({'sales':['sum', 'mean'], 'onpromotion' :['max']})
)
```

## Converters

```python
missing = lambda x: x if x not in [''] else 0
currency = lambda x: f"${x}.00"

oil = pd.read_csv('/content/oil2.csv',
                  names = ['date', 'oilprice','new'],
                  parse_dates= True,
                  header = None,
                  #skiprows=[0],
                  converters = {'oilprice':missing, 'new':currency})

oil
```

# Export files

```python
trans = pd.read_csv('/content/transactions.csv',
                    parse_dates=['Date'],
                    names = ['Date','Store_number', 'Transaction_count'],
                    skiprows = [0],
                    dtype = {'Store_number':'int8', 'Transaction_count':'int16'}).assign(
                    target_pct = lambda x: (x['Transaction_count']/2500),
                    met_target = lambda x: ((x['Transaction_count']/2500) >=1),
                    bonus_payable = lambda x: (((x['Transaction_count']/2500) >=1)*100),
                    month = lambda x: (x['Date'].dt.month),
                    day_of_week = lambda x: (x['Date'].dt.dayofweek)).astype({'target_pct':'Float32',
                                                                              'month':'int8',
                                                                              'day_of_week':'int8'})

trans.to_csv('cleaned.csv')
```
